```python
import pandas as pd
import re
import os

# 1. Fungsi untuk menghapus kata dengan huruf <= 3
def hapus_kata_pendek(data):
    """
    Menghapus kata-kata yang panjang hurufnya <= 3.
    Contoh:
    'aku di rumah' -> 'rumah'
    'ini bagus banget' -> 'bagus banget'
    """
    if not isinstance(data, str):
        return ''
    
    # Hapus tanda baca supaya tidak mengganggu perhitungan panjang huruf
    teks_bersih = re.sub(r'[^\w\s]', '', data)
    
    # Pisahkan jadi kata-kata
    daftar_kata = teks_bersih.split()
    
    # Ambil hanya kata dengan panjang > 3 huruf
    kata_yang_disaring = [kata for kata in daftar_kata if len(kata) > 3]
    
    # Gabungkan kembali jadi kalimat
    return ' '.join(kata_yang_disaring)

# --- 2. Proses DataFrame ---
try:
    # Path file input
    path_input = os.path.join('DatasetNew','Tahap3HapusDuplikat.csv')
    data = pd.read_csv(path_input)
    
    print("File CSV berhasil dibaca. Berikut 5 baris pertama data asli:")
    print(data.head())
    print("-" * 50)
    
    # Ganti dengan nama kolom data kamu
    nama_kolom_asli = 'full_text'
    
    # Terapkan fungsi pembersihan
    data['teks_bersih'] = data[nama_kolom_asli].apply(hapus_kata_pendek)
    
    print("Fungsi berhasil diterapkan. Berikut hasilnya:")
    print(data[[nama_kolom_asli, 'teks_bersih']].head())
    print("-" * 50)
    
    # Simpan hasilnya ke file baru
    folder_tujuan = 'DatasetNew'
    nama_file_output = 'Tahap4HapusKurang3Huruf.csv'
    path_output = os.path.join(folder_tujuan, nama_file_output)
    
    data.to_csv(path_output, index=False, encoding='utf-8')
    
    print(f"DataFrame yang sudah bersih berhasil disimpan di: '{path_output}'")

except FileNotFoundError:
    print("Error: File tidak ditemukan. Mohon periksa kembali path file input Anda.")
```

## Pra Pemrosesan teks/Pre processing text
Ada beberapa tahapan pra pemrosesn yang saya lakukan, tetapi karena kolom yang akan saya pakai hanya kolom isi ulasan X saja atau kolom yang bernama "full_text" maka saya akan menyimpan kolom tersebut kedalam satu file terlebih dahulu agar meringankan beban komputasi data.Setelah dilakukan penghapusan kata kurang dari 3 huruf didapatkan kolom baru yaitu kolom "teks_bersih", selanjutnya adalah proses preprocessing. Beberapa tahapan yang saya lakukan sebagai berikut:
1. casefolding data atau biasa disebut dengan pembersihan data seprti menyamakan hurup menjadi huruf kecil semua, menghapus karakter khusus seperti ?!@#$, dan menghapus angka.

```python
# Cek apakah ada missing value di dalam dataset
data.isnull().sum()
```

```python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory


#1. Casefolding data
def caseFolding(data, text_field, new_text_field_name):
    # Mengubah huruf kapital menjadi kecil
    data[new_text_field_name] = data[text_field].str.lower()
    # Menghapus karakter seperti @, #
    data[new_text_field_name] = data[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))
    # Menghapus angka
    data[new_text_field_name] = data[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem))
    return data

data_clean = caseFolding(data, 'teks_bersih', 'clean_text')
data_clean.head(10)
```

```python
# 2. Normalisasi kata tidak baku
def replace_taboo_words(text, kamus_tidak_baku):
    if isinstance(text, str):
        words = text.split()
        replaced_words = []
        kalimat_baku = []
        kata_diganti = []
        kata_tidak_baku_hash = []

        for word in words:
            if word in kamus_tidak_baku:
                baku_word = kamus_tidak_baku[word]
                if isinstance(baku_word, str) and all(char.isalpha() or char.isspace() for char in baku_word):
                    replaced_words.append(baku_word)
                    kalimat_baku.append(baku_word)
                    kata_diganti.append(word)
                    kata_tidak_baku_hash.append(hash(word))
                else:
                    replaced_words.append(word)
            else:
                replaced_words.append(word)

        replaced_text = ' '.join(replaced_words)
        return replaced_text, kalimat_baku, kata_diganti, kata_tidak_baku_hash
    else:
        return "", [], [], []

# Load kamus data
kamus_data = pd.read_excel("kamus\kamuskatabaku.xlsx")
kamus_tidak_baku = dict(zip(kamus_data["tidak_baku"], kamus_data["kata_baku"]))
results = data["clean_text"].apply(lambda x: replace_taboo_words(x, kamus_tidak_baku))
data["normalisasi_text"], data["kata_baku"], data["kata_tidak_baku"], data["kata_tidak_baku_Hash"] = zip(*results)
data[['full_text', 'teks_bersih', 'clean_text', 'normalisasi_text']].head(10)
```

```python
# 3. tokenizing (memisahkan teks menjadi potongan potongan)

def tokenize(text):
    tokens = text.split()
    return tokens

data["tokenize"] = data["normalisasi_text"].apply(tokenize)
data.head(10)
```

```python
#  4. Stopword removal (menghapus kata umum atau sambung contoh yang, di, dan)

# Unduh stopwords bahasa Indonesia
nltk.download('stopwords')

# Mendapatkan stopwords bahasa Indonesia
stop_words = stopwords.words('indonesian')
def remove_stopwords(text):
    return [word for word in text if word not in stop_words]

data['stopword_removal'] = data['tokenize'].apply(lambda x: remove_stopwords(x))
data.head(10)
```

```python
 # Step 5: Stemming

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stem_text(text):
    return [stemmer.stem(word) for word in text]

data['steming_data'] = data['stopword_removal'].apply(lambda x:' '.join(stem_text(x)))
data.head(10)
```

Melakukan pelabelan self training menggunakan sampel data berlabel manual 250 data

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# 1. Load data

df_labeled = pd.read_csv("LabelingData/Dataset250Labeled.csv")      # berlabel
df_unlabeled = pd.read_csv("LabelingData/Dataset750Unlabeled.csv")  # tanpa label

print("Jumlah data labeled  :", len(df_labeled))
print("Jumlah data unlabeled:", len(df_unlabeled))
```

```python
# 2. VALIDASI DAN PERBAIKAN KOLOM
if "steming_data" not in df_labeled.columns:
    raise ValueError("Kolom 'steming_data' tidak ditemukan pada Dataset250Labeled.csv!")

if "steming_data" not in df_unlabeled.columns:
    raise ValueError("Kolom 'steming_data' tidak ditemukan pada Dataset750Unlabeled.csv!")

# Hapus kolom label kosong pada dataset unlabeled
if "label" in df_unlabeled.columns:
    df_unlabeled = df_unlabeled.drop(columns=["label"])

# Hilangkan NaN
df_labeled["steming_data"] = df_labeled["steming_data"].fillna("")
df_unlabeled["steming_data"] = df_unlabeled["steming_data"].fillna("")
```

```python
# 3. SIAPKAN TRAINING DATA (250 data berlabel)
X_train_text = df_labeled["steming_data"]
y_train = df_labeled["label"]

# 4. TF-IDF VECTORIZER
vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    min_df=2,
    sublinear_tf=True
)

X_train = vectorizer.fit_transform(X_train_text)
```

```python
# 5. TRAIN MODEL
model = LogisticRegression(
    max_iter=3000,
    n_jobs=-1,
    class_weight="balanced"
)

model.fit(X_train, y_train)

print("\nModel berhasil dilatih menggunakan 250 data berlabel.")
```

```python
# 6. PREDIKSI DATA UNLABELED
X_unlabeled = vectorizer.transform(df_unlabeled["steming_data"])
proba = model.predict_proba(X_unlabeled)

# Label hasil prediksi
classes = model.classes_
confidence = np.max(proba, axis=1)
pred_label = classes[np.argmax(proba, axis=1)]

# Tambahkan ke dataframe
df_unlabeled["pseudo_label"] = pred_label
df_unlabeled["confidence"] = confidence
```

```python
# 7. FILTER CONFIDENCE >= 0.80 (Agar Tidak Kosong dan dapat disesuaikan)
threshold = 0.30
df_pseudo = df_unlabeled[df_unlabeled["confidence"] >= threshold].copy()

print("\nSummary Confidence:")
print("Max confidence :", df_unlabeled["confidence"].max())
print("Min confidence :", df_unlabeled["confidence"].min())
print("Mean confidence:", df_unlabeled["confidence"].mean())

print("\nJumlah pseudo-label terpilih (>= 0.70):", len(df_pseudo))
```

```python

# 8. GABUNGKAN DATA BERLABEL DAN PSEUDO LABEL
df_pseudo_final = df_pseudo[["steming_data", "pseudo_label"]] \
    .rename(columns={"pseudo_label": "label"})

df_final = pd.concat([df_labeled, df_pseudo_final], ignore_index=True)

print("Total data setelah digabung :", len(df_final))
```

```python

# 9. SIMPAN FILE HASIL
df_pseudo.to_csv("LabelingData/PseudoLabelsConf70.csv", index=False, encoding="utf-8")
df_final.to_csv("LabelingData/DatasetFinalSelfTraining.csv", index=False, encoding="utf-8")

print("\n--- PROSES SELESAI ---")
print("Pseudo label tersimpan di: LabelingData/PseudoLabelsConf30.csv")
print("Dataset final         di: LabelingData/DatasetFinalSelfTraining.csv")
```
